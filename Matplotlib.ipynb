```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(4,15,100)
y=np.sin(x)
```

```python
plt.figure(figsize=(8,6))
plt.title("Line plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.plot(x,y,color="red",label="Line")
plt.legend()
plt.show()
           
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
colors=np.random.rand(100)
size=1000*np.random.rand(100)
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter plot")
plt.xlabel("X values")
plt.ylabel("Y values")
# plt.grid(True)
plt.scatter(x,y,c=colors,s=size,cmap="plasma",alpha=0.7)
plt.colorbar(label="Intensity")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Scatter plot")
plt.xlabel("X values")
plt.ylabel("Y values")
# plt.grid(True)
plt.scatter(x,y,c=colors,s=size,cmap="viridis",alpha=0.6)
plt.show()
```

```python
data=np.random.rand(1000)
plt.hist(data,bins=20,color="Yellow",edgecolor="black",alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Histogram")
plt.xlabel("X values")
plt.ylabel("Y values")
# plt.grid(True)
plt.hist(data,bins=20,color="Red",edgecolor="black",alpha=0.7)
plt.show()
```

```python
data=np.random.normal(size=1000)
plt.figure(figsize=(5,3))
plt.title("Histogram")
plt.xlabel("X values")
plt.ylabel("Y values")
# plt.grid(True)
plt.hist(data,histtype="bar",bins=30,color="black",edgecolor="black",alpha=0.7)
plt.show()
```

```python
size=np.array([34,56,21,69])
label=np.array(["ram","sham","gita","sita"])
```

```python
plt.figure(figsize=(5,4))
plt.pie(size,labels=label,explode=[0.0,0.0,0.2,0.0],autopct="%1.1f%%",
        colors=["r","g","y","b"])
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.pie(size,labels=label,explode=[0.1,0.0,0.1,0.0],autopct="%1.1f%%",
        colors=["r","g","y","b"])
plt.title("pie chart")
plt.axis("equal")
plt.show()
```

```python
plt.figure(figsize=(5,4))
cols=["cyan","teal","turquoise","skyblue"]
plt.pie(size,labels=label,explode=[0.0,0.0,0.1,0.0],autopct="%1.1f%%",
        colors=cols)
plt.title("pie chart")
plt.axis("equal")
plt.show()
```

```python
angle=np.linspace(0,2*np.pi,100)
redii=np.random.rand(100)
```

```python
plt.figure(figsize=(6,5))
ax=plt.subplot(111,polar=True)
ax.plot(angle,redii,color="red",linewidth=2)
ax.fill(angle,redii,color="red",alpha=0.3)
plt.title("polar plot")
plt.show()
```
